# INSTRUCCION SQL

Uniendo comandos + clausulas + operadores + funciones = instrucción SQL

## TIPOS DE COMANDOS
DDL: Data definition languague, para crear y modificar estructura de base de datos
DML: Data manipulation languague, seleccionar registros de base de datos, también para insertar nuevos registros, actualizar, borrar info
DCL: Data control language, brindar seguridad
TCL: Transaction control languague, para gestión de cambios en los datos

- DDL:
CREATE
ALTER
DROP
TRUNCATE

- DML:
SELECT
INSERT
UPDATE
DELETE

- DCL:
GRANT
REVOKE

- TCL
COMMIT
ROLLBACK 
SAVEPOINT

## CLAUSULAS:
FROM
WHERE
GROUP BY
HAVING 
ORDER BY

Nota: Obligatorio sera comando y clausula, los operadores y funciones no son estricamente necesarios
