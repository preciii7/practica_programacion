## Importacion de los datos y librerias

```python
## Conectarse a docker y pandas
import pandas as pd
import os
from sqlalchemy import create_engine
```

## Creacion usuario para conectarse a base de datos

```python
user = os.getenv("MYSQL_USER")
pwd  = os.environ["MYSQL_PASSWORD"]
host = os.getenv("MYSQL_HOST")
port = os.getenv("MYSQL_PORT")
db   = os.getenv("MYSQL_DB")

engine = create_engine(f"mysql+pymysql://{user}:{pwd}@{host}:{port}/{db}")
```

```python
pd.read_sql("SELECT rating, COUNT(film_id) AS frecuencia FROM film GROUP BY rating ORDER BY frecuencia DESC",engine)
```
