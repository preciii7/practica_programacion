## Importacion de librerias y acceso a contenedor sql

```python
import os
import pandas as pd
from dotenv import load_dotenv
from sqlalchemy import create_engine

load_dotenv("credenciales/practica.env")

host = os.getenv("MYSQL_HOST", "127.0.0.1")
port = os.getenv("MYSQL_PORT", "3306")
db = os.getenv("MYSQL_DB", "practica_ejercicios")

user = os.getenv("MYSQL_USER")
pwd  = os.getenv("MYSQL_PASSWORD")  # SOLO esta


engine = create_engine(f"mysql+pymysql://{user}:{pwd}@{host}:{port}/{db}")
```

### Consultas de agrupacion 

#### Funciones de agregado de SQL
- AVG: PROMEDIO
- COUNT: CONTADOR DE REGISTROS
- SUM: SUMA
- MAX: MAXIMO
- MIN: MINIMO

Necesitamos dos campos, uno para agrupar y otro para operar, por eso utilizamos el comando GROUP BY



```python
pd.read_sql("SELECT SECCIÓN, SUM(PRECIO) FROM productos GROUP BY SECCIÓN;",engine)
```

#### Nota: Se puede crear un alias

```python
pd.read_sql("SELECT SECCIÓN, SUM(PRECIO) AS SUMA_ARTICULOS FROM productos GROUP BY SECCIÓN ORDER BY SUMA_ARTICULOS;",engine)
```

```python
# porque group by y porque no funciona asc?
pd.read_sql("SELECT SECCIÓN, AVG(PRECIO) AS MEDIA_ARTICULOS FROM productos GROUP BY SECCIÓN HAVING SECCIÓN = 'CERÁMICA' OR SECCIÓN = 'DEPORTES';",engine)
```

### Consultas de calculo
Cada gestor de base de datos tiene las suyas propias
- Now(): Dia y hora actuales
- Datediff(): Diferencia que hay entre dos fechas
- Date_format(): Para quitar decimales y redondear
- Round():
- Truncate():
- Concat(): 

```python
pd.read_sql("SELECT NOMBREARTÍCULO, SECCIÓN, PRECIO, ROUND(PRECIO+(PRECIO*0.21),2) AS PRECIO_IVA FROM productos",engine)
```

```python
"""
1. Realizar una consulta que visualice los campos NOMBRE ARTÍCULO,
SECCIÓN, PRECIO de la tabla PRODUCTOS y un campo nuevo que
nombramos con el texto “DESCUENTO_7”. Debe mostrar el resultado de aplicar
sobre el campo PRECIO un descuento de un 7 %. El formato del nuevo campo
para debe aparecer con 2 lugares decimales. 
"""
pd.read_sql("SELECT NOMBREARTÍCULO, SECCIÓN, PRECIO, ROUND(PRECIO-(PRECIO*0.07),2) AS DESCUENTO_7 FROM productos ",engine)
```

```python
# Utilizando funciones NOW() y DATEDIFF()
pd.read_sql("SELECT FECHA, NOW(), DATEDIFF(NOW(),FECHA) AS DIAS_DIFERENCIA FROM productos",engine)
```

```python
"""
2.Realizar una consulta visualizando los campos FECHA, SECCIÓN, NOMBRE
ARTÍCULO y PRECIO de la tabla PRODUCTOS y un campo nuevo que
nombramos con el texto “DTO2 €_EN_CERÁMICA”. Debe mostrar el resultado
de aplicar sobre el campo PRECIO la resta de 2 € sólo a los artículos de la sección
CERÁMICA. El formato del nuevo campo debe aparecer con 2 lugares decimales.
Ordenar el resultado de la consulta por el campo FECHA descendente. 
"""
pd.read_sql("SELECT FECHA, SECCIÓN, NOMBREARTÍCULO, PRECIO, PRECIO-2 AS DTO2e_EN_CERAMICA FROM productos",engine)
```

```python
""" 
Realizar una consulta visualizando los campos FECHA, SECCIÓN, NOMBRE
ARTÍCULO y PRECIO de la tabla PRODUCTOS y un campo nuevo que
nombramos con el texto “DTO2 €_EN_CERÁMICA”. Debe mostrar el resultado
de aplicar sobre el campo PRECIO la resta de 2 € sólo a los artículos de la sección
CERÁMICA. El formato del nuevo campo debe aparecer con 2 lugares decimales.
Ordenar el resultado de la consulta por el campo FECHA descendente. 
"""
pd.read_sql("SELECT FECHA, SECCIÓN, NOMBREARTÍCULO, PRECIO, ROUND(PRECIO-2,2) AS DTO2e_EN_CERAMICA FROM productos WHERE SECCIÓN = 'CERAMICA' ORDER BY FECHA DESC",engine)
```

```python
""" 
Realizar una consulta visualizando los campos NOMBRE ARTÍCULO, SECCIÓN, PRECIO de la tabla PRODUCTOS y
un campo nuevo que nombramos con el texto “PRECIO_AUMENTADO_EN_2”. 
Debe mostrar el PRECIO con un incremento de un 2% del PRECIO. 
Sólo debemos tener en cuenta los registros de la sección FERRETERÍA. 
El nuevo campo debe aparecer en Euros y con 2 lugares decimales.
"""
pd.read_sql("SELECT NOMBREARTÍCULO, SECCIÓN, PRECIO, ROUND(PRECIO*1.02,2) AS PRECIO_AUMENTADO_EN_2  FROM productos WHERE SECCIÓN = 'FERRETERÍA'", engine)
```
