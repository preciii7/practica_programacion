### Comandos SQL
- DDL: (Data Definition Language o Lenguaje de Definición de Datos), en PostgreSQL se utilizan para definir, modificar y eliminar la estructura de los objetos de la base de datos, como tablas, índices y esquemas. PostgreSQL destaca porque sus comandos DDL son transaccionales, lo que significa que puedes ejecutarlos dentro de un bloque BEGIN ... COMMIT y revertirlos con ROLLBACK si algo falla.
- DML: (Data Manipulation Language o Lenguaje de Manipulación de Datos) en PostgreSQL se utilizan para gestionar los datos que viven dentro de las tablas. A diferencia del DDL (que cambia la estructura), el DML afecta a las filas de información.
- DCL: (Data Control Language o Lenguaje de Control de Datos) se utilizan para gestionar la seguridad de la base de datos, permitiendo a los administradores otorgar o retirar permisos a los usuarios y roles sobre objetos específicos.
- TCL: (Transaction Control Language o Lenguaje de Control de Transacciones) se utilizan para gestionar los cambios realizados por los comandos DML, Su función principal es agrupar varias operaciones en una sola unidad de trabajo para asegurar la integridad de los datos, En PostgreSQL, si no abres una transacción explícita, cada comando se ejecuta en modo "autocommit" (se guarda automáticamente).

Para DDL
- CREATE: Se utiliza para crear nuevos objetos (tablas, índices, vistas, funciones, etc.)

### Consultas Multitabla
#### Union externa
- UNION: (Hay restricciones) Unir dos tablas pero se requiere que estas dos tengan el mismo numero de columnas (variables) y tambien el tipo de las variables (cuantitativa o cualitativa) debe ser semejante, el nombre no importa
- UNION ALL: En UNION puede haber registros repetidos asi que solo muestra ese registro solo una vez, con UNION ALL se muestran todos los registros repetidos

Tambien se pueden unir mas de dos tablas

### Subconsultas
Son consultas dentro de otras consultas
- Escalonada
- de lista
- correlacionada

```python
# Subconsulta escalonada
pd.read_sql("SELECT SECCIÓN, NOMBREARTÍCULO FROM productos WHERE PRECIO > (SELECT AVG(PRECIO) FROM productos)",engine)
```

```python
# Subconsulta de lista
pd.read_sql("(SELECT PRECIO FROM productos WHERE SECCIÓN = 'CERAMICA')",engine)
```

```python
pd.read_sql("SELECT * FROM productos WHERE PRECIO > ALL (SELECT PRECIO FROM productos WHERE SECCIÓN = 'CERAMICA')",engine)
```

```python
# Si cambiamos el ALL por el ANY entonces quiere decir que busca cualquier producto que tenga precio mayor a uno de los de la lista
pd.read_sql("SELECT * FROM productos WHERE PRECIO > ANY (SELECT PRECIO FROM productos WHERE SECCIÓN = 'CERAMICA')",engine)
```

```python
pd.read_sql("SELECT * FROM productos WHERE PRECIO > ALL (SELECT PRECIO FROM productos WHERE SECCIÓN = 'JUGUETERÍA')",engine)
```

```python
pd.read_sql("SELECT * FROM productos WHERE PRECIO > ANY (SELECT PRECIO FROM productos WHERE SECCIÓN = 'JUGUETERÍA')",engine)
```

```python
pd.read_sql("SELECT NOMBREARTÍCULO,PRECIO FROM productos_access WHERE CÓDIGOARTÍCULO IN (SELECT `CÓDIGO ARTÍCULO` AS codigo FROM productos_pedidos_access WHERE UNIDADES > 20)",engine)
```
