```python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad=edad
    def saludar(self):
        return f"Hola {self.nombre}"

cliente = Persona("Pepito", "30")

print(cliente.saludar())
```

```python
class Perro():
    def __init__ (self):
        self.nombre = "Max"
        self.raza = "ladrador"
    def ladrar(self):
        print(f"{self.nombre} esta ladrando")

Perro1=Perro()
Perro1.ladrar()
```

```python
class Libro():
    def __init__(self, autor, titulo, año_publicacion):
        self.autor = autor
        self.titulo = titulo
        self.año_publicacion = año_publicacion
    def mostrar_info(self):
        return f"{self.autor}, {self.titulo}, {self.año_publicacion}"

libro1=Libro("Michael", "El señor de los anillos", 1991)
libro1.mostrar_info()
```

```python
class Rectangulo():
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura
    def area(self):
        return self.base * self.altura
    def perimetro(self):
        return 2*(self.base + self.altura)

rectangulo1 = Rectangulo(2.1,3.3)
print("area: ", rectangulo1.area())
print("perimetro: ", rectangulo1.perimetro())
```
