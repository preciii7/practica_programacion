## Nivel 1 basico

```python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad=edad
    def saludar(self):
        return f"Hola {self.nombre}"

cliente = Persona("Pepito", "30")

print(cliente.saludar())
```

```python
class Perro():
    def __init__ (self):
        self.nombre = "Max"
        self.raza = "ladrador"
    def ladrar(self):
        print(f"{self.nombre} esta ladrando")

Perro1=Perro()
Perro1.ladrar()
```

```python
class Libro():
    def __init__(self, autor, titulo, año_publicacion):
        self.autor = autor
        self.titulo = titulo
        self.año_publicacion = año_publicacion
    def mostrar_info(self):
        return f"{self.autor}, {self.titulo}, {self.año_publicacion}"

libro1=Libro("Michael", "El señor de los anillos", 1991)
libro1.mostrar_info()
```

## Nivel 2

```python
class Rectangulo():
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura
    def area(self):
        return self.base * self.altura
    def perimetro(self):
        return 2*(self.base + self.altura)

rectangulo1 = Rectangulo(2.1,3.3)
print("area: ", rectangulo1.area())
print("perimetro: ", rectangulo1.perimetro())
```

```python
class CuentaBancaria():
    def __init__(self, titular, saldo):
        self.titular = titular
        self.saldo = saldo
    def depositar(self, cantidad):
        self.saldo+=cantidad
    def retirar(self,cantidad):
        if self.saldo - cantidad >=0:
            self.saldo -=cantidad
        else:
            print("Fondos insuficientes")
    def mostrar_saldo(self):
        return self.saldo

cuenta1= CuentaBancaria("pepito", 10000)
cuenta1.depositar(2000)
print(cuenta1.mostrar_saldo())
cuenta1.retirar(13000)
print(cuenta1.mostrar_saldo())
```

```python
class Coche():
    def __init__(self,marca,modelo):
        self.marca = marca
        self.modelo = modelo
        self.velocidad = 0
    def acelerar(self, cantidad):
        self.velocidad +=cantidad
    def frenar(self,cantidad):
        if self.velocidad - cantidad >=0:
            self.velocidad -=cantidad
        else:
            print("La velocidad no puede ser negativa")
        
ferrari = Coche("ferrari", "mustang")
ferrari.acelerar(1000)
print(vars(ferrari))
ferrari.frenar(500)
print(vars(ferrari))
```

```python
class Lampara():
    def __init__(self, marca):
        self.marca = marca
        self.encendida = False
    def encender(self):
        self.encender = True

    def apagar(self):
        self.encender = False
    
    def estado(self):
        if self.encender == True:
            print("La lampara esta encendida")
        if self.encender == False:
            print("La lampara esta apagada")

lampara1=Lampara("LG")
lampara1.encender()
lampara1.estado()
lampara1.apagar()
lampara1.estado()
```

```python
class EstudianteNotas():
    def __init__(self, nombre):
        self.nombre = nombre
        self.notas = []
    
    def agregar_nota(self, nota):
        self.notas.append(nota)
    
    def promedio(self):
        return sum(self.notas)/len(self.notas)
    
    def aprobo(self):
        if float(self.promedio()) > 3.0:
            print("Aprobo")
        else:
            print("No aprobo")

notasPepito = EstudianteNotas("Pepito")
notasPepito.agregar_nota(3)
notasPepito.agregar_nota(4)
notasPepito.agregar_nota(5)

print(notasPepito.promedio())

notasPepito.aprobo()
```

```python
class CarritoCompras():
    def __init__(self):
        self.productos = []
    def agregar_producto(self, producto):
        self.productos.append(producto)
    def mostrar_productos(self):
        return self.productos
    def cantidad_productos(self):
        return len(self.productos)

carrito1= CarritoCompras()
carrito1.agregar_producto(1)
print(carrito1.mostrar_productos())
carrito1.agregar_producto(2)
print(carrito1.cantidad_productos())
```
