## Nivel 1: básicos

```python
#Crea una función que imprima "Hola mundo".
fun1 = lambda a: print(a)

(fun1("hello world"))
```

```python
#Crea una función que reciba un nombre y salude a esa persona.
fun2 = lambda a: print(f"Hello {a}")

fun2("Pepito")
```

```python
#Crea una función que reciba dos números y devuelva su suma.
fun3 = lambda a,b: a+b

fun3(1,2)
```

```python
#Crea una función que reciba dos números y devuelva el mayor.
fun4 = lambda a,b: print(a) if a>b  else print(b)

fun4(1,2)
```

```python
# Crea una función que reciba un número y devuelva True si es par y False si es impar.

fun5 = lambda a: True if a %2 == 0 else False

fun5(7)
```
